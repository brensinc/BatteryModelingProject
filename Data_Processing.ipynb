```python
exper_data1 = pd.read_excel("12_2_2015_Incremental OCV test_SP20-1.xlsx", sheet_name = "Channel_1-005_1")
exper_data2 = pd.read_excel("12_2_2015_Incremental OCV test_SP20-1.xlsx", sheet_name = "Channel_1-005_2")
exper_data3 = pd.read_excel("12_2_2015_Incremental OCV test_SP20-1.xlsx", sheet_name = "Channel_1-005_3")
```

```python
exper_data = pd.concat([exper_data1, exper_data2, exper_data3], ignore_index=True)
exper_data.shape
```

```python
C_max = 2
```

```python
def compute_soc(row):
    return (C_max - abs(row['Discharge_Capacity(Ah)'] - row['Charge_Capacity(Ah)'])) / C_max 

exper_data['SOC'] = exper_data.apply(compute_soc, axis=1)
```

```python
exper_data["SOC"] = exper_data["SOC"].round(2)
exper_data = exper_data.groupby("SOC").agg("mean").reset_index()
exper_data
```

```python
exper_data.to_csv("preprocessed_battery_data.csv", index=False)
```
